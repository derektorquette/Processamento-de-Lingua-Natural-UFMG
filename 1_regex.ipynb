## AGRUPAMENTO
#COMANDO - .compile e .finditer

image.png

```python
import re

amostras = """Hahahahhahaha
huahuahuahuahuahua
haha
kkkk
hahahahahaha"""

regex = re.compile(r"k+|([Hh]a)+|(Hh]ua)+")
interador = regex.finditer(amostras)

for ex in interador:
  inicio, fim = ex.span()
  print(inicio, fim, amostras[inicio:fim])
```

##CARACTERES ESPECIAIS

image.png



```python
import re

def preprocess_text(text):
    # Remover pontuações e tokenização
    tokens = re.findall(r'\b\w+\b', text)
    # Normalização para minúsculas e remoção de espaços em branco
    tokens = [token.lower() for token in tokens if token.strip()]
    return tokens

texto = "Olá! Este é um exemplo de pré-processamento de texto, com pontuações e maiúsculas. Vamos removê-las!"
tokens_preprocessados = preprocess_text(texto)
print(tokens_preprocessados)
```
