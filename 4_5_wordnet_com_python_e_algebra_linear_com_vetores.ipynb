## **Matriz de frequência Termo-documento** Dado um vocabulário e um conjunto de documentos, as representações das palvras e dos documentos podem ser calculadas a partir da contagem de cada palavra em cada documento

```python
import nltk
nltk.download('punkt')
from sklearn.feature_extraction.text import CountVectorizer


def tokenize(texto):
  return nltk.word_tokenize(texto)

corpus = ['Chove chuva',
'Chove sem parar',
'Chove chuva',
'Chove sem parar']

vectorizer = CountVectorizer(tokenizer=tokenize)

vetores = vectorizer.fit_transform(corpus)

vetores.toarray()

vocab = vectorizer.get_feature_names_out()
vetores.toarray(), vocab
```

##**MATRIZ TERMO-TERMO**

```python
corpus = ['Pra Deus, nosso Senhor',
 'Pra chuva parar']

corpus_tok = [verso.split() for verso in corpus]

vocab = ["Pra", "Deus", "nosso", "Senhor", "chuva", "parar"]
vetores = np.zeros((len(vocab), len(vocab)))

for verso in corpus_tok:
  for i, w1 in enumerate(vocab):
    for j, w2 in enumerate(vocab):
      if i != j:
       if w1 in verso and w2 in verso:
         vetores[i, j] += 1

print('Vocabulário')
print(vocab)
print()
print('Matrix')
print(vetores)
```
